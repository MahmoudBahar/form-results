```python
%%writefile form.py
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import streamlit as st
import numpy as np
import psycopg2
import re

if 'server' not in globals():
    server = 'mtnatega.postgres.database.azure.com'
    database = 'postgres'
    username = 'natega'
    password = 'Mt100323'
    port = '5432'
    sslmode = 'require'

@st.dialog("مسح بياناتك")
def delete():
    st.session_state['success'] = False
    st.session_state['message'] = ''
    st.write(f"هل انت متأكد انك عاوز تمسح بياناتك؟")
    reason = st.text_input("الرقم القومي", key='ID_delete', placeholder='الرقم القومي', label_visibility='collapsed',max_chars = 14)
    if st.button("تأكيد", use_container_width=True):
        if not st.session_state['ID_delete'].isalnum() or not len(st.session_state['ID_delete']) == 14:
            st.toast('الرقم القومي ليس صحيح', icon='🚨')
        while True:
            try:
                conn = psycopg2.connect(
                    host=server,
                    database=database,
                    user=username,
                    password=password,
                    port=port,
                    sslmode=sslmode
                )
                cur = conn.cursor()
                select_query = '''
                    SELECT id FROM form
                    '''
                cur.execute(select_query)
                if st.session_state['ID_delete'] not in tuple(i[0] for i in cur.fetchall()):
                    st.toast('الرقم القومي ليس مسجل', icon='🚨')
                    break
                cur = conn.cursor()
                delete_query = '''
                    DELETE FROM form
                    WHERE id = %s
                    '''
                cur.execute(delete_query, (st.session_state['ID_delete'],))
                conn.commit()
                cur.close()
                conn.close()
                st.toast('تم المسح', icon='🗑️')
                break
            except:
                pass


if "number" not in globals():
    conn = psycopg2.connect(
    host=server,
    database=database,
    user=username,
    password=password,
    port=port,
    sslmode=sslmode
    )
    cur = conn.cursor()
    select_query = '''
        SELECT add_number FROM form
        '''
    cur.execute(select_query)
    data = pd.DataFrame(cur.fetchall()).value_counts()
    sums = 0
    for i, j in enumerate(data, start = 1):
        sums += j*i
    number = sums
    cur.close()
    conn.close()

st.set_page_config(page_title='زيادة مرافقين', page_icon='🎓', initial_sidebar_state='collapsed')
if 'message' not in st.session_state:
    st.session_state['message'] = '' 

def is_arabic_and_spaces_only(text):
    # Regular expression to match Arabic characters and spaces
    arabic_and_space_pattern = re.compile(r'^[\u0600-\u06FF\s]+$')

    # Check if the text matches the pattern
    return bool(arabic_and_space_pattern.match(text))

# Initialize session state for storing the result
if 'dep' not in st.session_state:
    st.session_state['dep'] = None
    st.session_state['name'] = None
    st.session_state['phone'] = None
    st.session_state['ID'] = None
    st.session_state['payment_no'] = None
    st.session_state['success'] = False

# Function to fetch data
  
def save_data():
    while True:
        try:
            conn = psycopg2.connect(
                host=server,
                database=database,
                user=username,
                password=password,
                port=port,
                sslmode=sslmode
            )
            cur = conn.cursor()
            select_query = '''
                SELECT add_number FROM form
                '''
            cur.execute(select_query)
            data = pd.DataFrame(cur.fetchall()).value_counts()
            sums = 0
            for i, j in enumerate(data, start = 1):
                sums += j*i
            number = sums
            cur = conn.cursor()
            select_query = '''
            SELECT id, pay_no FROM form
            '''
            cur.execute(select_query)
            ids = tuple(item for sublist in cur.fetchall() for item in sublist)
            if number < 400:
                if ids.count(st.session_state['ID']) == 0:
                    if st.session_state['pay_no'] not in ids:
                        insert_query = '''
                        INSERT INTO form(name, dep, id, phone, pay_no, add_number, id_image_name, id_image, pay_image_name, pay_image) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
                        '''
                        cur.execute(insert_query, (st.session_state['name'], st.session_state['dep'], st.session_state['ID'],
                                                   st.session_state['phone'], st.session_state['pay_no'], st.session_state['add_no'],
                                                   st.session_state['ID_image'].name,st.session_state['ID_image'].getvalue(),
                                                   st.session_state['pay_image'].name,st.session_state['pay_image'].getvalue()
                                                  ))
                        conn.commit()
                        st.session_state['success'] = True
                    else:
                        st.session_state['message'] = 'رقم التحويل مكرر'
                else:
                    st.session_state['message'] = 'انت سجلت بالفعل'
            else:
                st.session_state['message'] = "العدد اكتمل"
            cur.close()
            conn.close()
            break
        except:
            pass

def handle_button():
    st.session_state['success'] = False
    st.session_state['message'] = ''
    if st.session_state['name'] == None:
        st.session_state['message'] = 'اكتب الاسم'
    elif st.session_state['dep'] == None:
        st.session_state['message'] = 'اختار القسم'
    elif st.session_state['ID'] == None:
        st.session_state['message'] = 'اكتب الرقم القومي'
    elif st.session_state['phone'] == None:
        st.session_state['message'] = 'اكتب رقم الهاتف'
    elif st.session_state['pay_no'] == None:
        st.session_state['message'] = 'اكتب رقم عملية الدفع'
    elif st.session_state['ID_image'] == None:
        st.session_state['message'] = 'ارفع صورة البطاقة الشخصية'
    elif st.session_state['pay_image'] == None:
        st.session_state['message'] = 'ارفع صورة التحويل'
    else:
        if not is_arabic_and_spaces_only(st.session_state['name']):
            st.session_state['message'] = 'اكتب الاسم بالعربي'
        elif len(st.session_state['name'].strip().split()) < 4:
            st.session_state['message'] = 'اكتب الاسم رباعي'
        elif len(st.session_state['ID']) != 14:
            st.session_state['message'] = 'اكتب رقم قومي صحيح'
        elif len(st.session_state['phone']) < 11:
            st.session_state['message'] = 'اكتب رقم هاتف صحيح'
        elif len(st.session_state['pay_no']) > 200:
            st.session_state['message'] = 'رقم دفع صحيح'
        else:
            save_data()

if number < 400:
    st.markdown("""
        <style>
        div.stButton > button:first-child {
            background-color: white;
            color: black
        }
        </style>""", unsafe_allow_html=True)
    st.button("مسح سجلاتي", use_container_width=True, on_click=delete)
    st.header(f"فورم زيادة المرافقين", divider='rainbow')
    with st.form("my_form"):
        col1, col2, col3 = st.columns([1.5, 1, 1])
        with col1:
            st.text_input('الاسم بالعربي رباعي', key='name', label_visibility='collapsed', placeholder="الاسم بالعربي رباعي")
        with col2:
            st.selectbox('القسم', key='dep', options=['ميكانيكا صناعية', 'ميكاترونيكس', 'اتصالات', 'كهرباء', 'حاسبات', 'ميكانيكا انتاج', 'اتصالات credit', 'باور credit', 'طبية'], label_visibility='collapsed', placeholder="القسم")
        with col3:
            st.selectbox('عدد المرافقين', key='add_no', options=['مرافق واحد (100 ج)', 'اتنين مرافقين (200 ج)', 'ثلاث مرافقين (300 ج)'], label_visibility='collapsed', placeholder="عدد المرافقين")
        col1, col2, col3 = st.columns([1, 1, 1])
        with col1:
            st.text_input('الرقم القومي', key='ID', label_visibility='collapsed', placeholder="الرقم القومي", max_chars = 14)
        with col2:
            st.text_input('رقم الهاتف', key='phone', label_visibility='collapsed', placeholder="رقم الهاتف")
        with col3:
            st.text_input('رقم عملية الدفع', key='pay_no', label_visibility='collapsed', placeholder="رقم عملية الدفع")
        col1, col2 = st.columns([1, 1])
        with col1:
            st.file_uploader('صورة البطاقة الشخصية', key='ID_image', type=['png', 'jpg'])
        with col2:
            st.file_uploader('صورة التحويل', key='pay_image', type=['png', 'jpg'])
        st.markdown("""
        <style>
        div.stButton > button:first-child {
            background-color: white;
            color: black
        }
        </style>""", unsafe_allow_html=True)
        st.form_submit_button("تأكيد", use_container_width=True, on_click=handle_button)
    if st.session_state['message'] != '':
        st.error(st.session_state['message'])
    if st.session_state['success']:
        st.success('تمت العملية بنجاح')
        st.balloons()
else:
    st.header(f"العدد اكتمل", divider='rainbow')
```

```python
import os, shutil

try:
    directory_path = 'ID/'
    shutil.rmtree(directory_path)
except Exception as e:
    directory_path = 'ID/'
    os.makedirs(directory_path, exist_ok=True)
try:
    directory_path = './PAY/'
    shutil.rmtree(directory_path)
except Exception as e:
    directory_path = './PAY/'
    os.makedirs(directory_path, exist_ok=True)
```

```python
for row in data:
    no, name, dep, id, phone, pay_no, add_no, id_image_name, id_image, pay_image_name, pay_image = row
    # Convert memoryview to bytes
    id_image_bytes = id_image.tobytes() if isinstance(id_image, memoryview) else id_image
    pay_image_bytes = pay_image.tobytes() if isinstance(pay_image, memoryview) else pay_image
    
    # Save ID image
    with open("./ID/"+name+'_'+id+ '_id' + id_image_name[id_image_name.rfind('.'):], 'wb') as id_image_file:
        id_image_file.write(id_image_bytes)
    
    # Save Pay image
    with open("./PAY/"+name+'_'+id+ '_pay' + pay_image_name[pay_image_name.rfind('.'):], 'wb') as pay_image_file:
        pay_image_file.write(pay_image_bytes)
```
